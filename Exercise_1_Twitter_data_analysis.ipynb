```python
import csv
import pandas as pd
twitter = []
with open('C:\\Users\\karthi\\Desktop\\twitter1.csv' , 'r') as f:
    csv_reader = csv.reader(f)
 
    for line in csv_reader:
        twitter.extend(line)
twitter
```

```python
import nltk
nltk.download('punkt')
```

```python
#Tokenization
import re
tokens = []
for each in twitter:
    var = re.match(r'#[a-zA-Z0-9]*',each)
    tokens.extend(var[0][1:].split(" "))
tokens
```

```python
from nltk import PorterStemmer
ps = PorterStemmer()
#ls = LancasterStemmer()
lis1 = []  
for i in twitter:
  lis1.append(nltk.word_tokenize(i)[1])
lis1
```

```python
print(" Porter's Stemmer\n\n")
temp = []
for each in lis1:
    temp.append([each , ps.stem(each)])
print(pd.DataFrame(temp))
```

```python
from nltk import SnowballStemmer
snow = SnowballStemmer("english")
print(" Snowball Stemmer \n\n")
temp = []
for each in lis1:
    temp.append([each , snow.stem(each)])
print(pd.DataFrame(temp))
```

```python
from nltk import LancasterStemmer
lancaster=LancasterStemmer()
print("  Lancaster Stemmer \n\n")
temp = []
for each in lis1:
    temp.append([each , lancaster.stem(each)])
print(pd.DataFrame(temp))
```

```python
print("Bigrams")
import collections
from nltk.util import ngrams
Bigrams = ngrams(lis1, 2)
Freq = collections.Counter(Bigrams)
t = []
for key in Freq.keys():
    t.append([key , Freq[key]])
print(pd.DataFrame(t))
```

```python
print("TriGrams")
tri = ngrams(lis1, 3)
Freq = collections.Counter(tri)
t = []
for key in Freq.keys():
    t.append([key , Freq[key]])
print(pd.DataFrame(t))
```

```python

from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True,ngram_range = (1,1),tokenizer = token.tokenize)
text_counts = cv.fit_transform(twitter)
```

```python
print("Doc term matrix")
df = pd.DataFrame(text_counts.toarray(), columns=cv.get_feature_names())
df
```
